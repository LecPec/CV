```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
omp = np.genfromtxt('data_omp.txt')
ordinary = np.genfromtxt('data.txt')
```

```python
fig, ax = plt.subplots()
ax.plot(omp[:, 0], ordinary[:, 1] / omp[:, 1], label='acceleration')
ax.set_xlabel('Space dimmension')
ax.set_ylabel('t, s')
ax.legend()
ax.set_title('Task 2')
fig.savefig('plot_acceleration_task2.png')
```

```python
omp = np.genfromtxt('data_cores.txt')
ordinary = np.genfromtxt('data_cores_posl.txt')

y = []

for i in range(2, 7):
    y.append(ordinary[0, 0] / omp[i, 1] / i)

fig, ax = plt.subplots()
ax.plot([i for i in range(2, 7)], y, label='effectiveness') # 8 cores on MacBook Air 2020 M1?????
ax.set_xlabel('N of threads')
ax.set_ylabel('t, s')
ax.legend()
ax.set_title('Task 2')
fig.savefig('plot_effectiveness_task2.png')
```

```python
omp = np.genfromtxt('data_omp_int.txt')
ordinary = np.genfromtxt('data_int.txt')
```

```python
fig, ax = plt.subplots()
ax.plot(omp[:len(ordinary), 0], ordinary[:, 1] / omp[:len(ordinary), 1], label='acceleration')
ax.plot(omp[:len(ordinary), 0], ordinary[:, 1] / omp[:len(ordinary), 1] / 8, label='effectivenes') # 8 cores on MacBook Air 2020 M1?????
ax.set_xlabel('h')
ax.set_ylabel('t, s')
ax.legend()
ax.set_title('Task 3')
fig.savefig('plot_acceleration_task3.png')
```
