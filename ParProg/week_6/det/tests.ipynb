# Ускорение в определителе

```python
posl = np.genfromtxt('determinant_data_from_dim_posl.txt')
par = np.genfromtxt('determinant_data_from_dim_parallel.txt')

fig, ax = plt.subplots()

ax.plot(par[:, 0], par[:, 1]/posl[:, 1], label='acceleration')
ax.set_xlabel('dim of space')

ax.legend()
```
