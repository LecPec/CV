```python
init_electrons = np.genfromtxt('../bigger_ptcls/1/els_curr.txt')
init_ions = np.genfromtxt('../bigger_ptcls/1/ions_curr.txt')

electrons = np.genfromtxt('1e_13/els_curr.txt')
ions = np.genfromtxt('1e_13/ions_curr.txt')

electrons_ = np.genfromtxt('2e_13/els_curr.txt')
ions_ = np.genfromtxt('2e_13/ions_curr.txt')

electrons__ = np.genfromtxt('5e_13/els_curr.txt')
ions__ = np.genfromtxt('5e_13/ions_curr.txt')
print(electrons__.shape, ions__.shape)

electrons___ = np.genfromtxt('5e_14/els_curr.txt')
ions___ = np.genfromtxt('5e_14/ions_curr.txt')

e = 1.6e-19

time = np.concatenate([init_electrons[:7000, 0]*1e-3, electrons[:, 0]*1e-4 + 7e6*1e-3])
timei = np.concatenate([init_ions[:7000, 0]*1e-3, ions[:, 0]*1e-4 + 7e6*1e-3])
time_ = np.concatenate([init_electrons[:7000, 0]*1e-3, electrons_[:, 0]*2*1e-4 + 7e6*1e-3])
time_i = np.concatenate([init_ions[:7000, 0]*1e-3, ions_[:, 0]*2*1e-4 + 7e6*1e-3])
time__ = np.concatenate([init_electrons[:7000, 0]*1e-3, electrons__[:, 0]*5*1e-4 + 7e6*1e-3])
time__i = np.concatenate([init_ions[:7000, 0]*1e-3, ions__[:, 0]*5*1e-4 + 7e6*1e-3])
time___ = np.concatenate([init_electrons[:7000, 0]*1e-3, electrons___[:, 0]*5*1e-5 + 7e6*1e-3])
time___i = np.concatenate([init_ions[:7000, 0]*1e-3, ions___[:, 0]*5*1e-5 + 7e6*1e-3])

electrons_curr = np.concatenate([init_electrons[:7000, 1]*e*1.5708e6/1e-9, electrons[:, 1]*e*1.5708e6/1e-10])
ions_curr = np.concatenate([init_ions[:7000, 1]*e*1.5708e6/1e-9, ions[:, 1]*e*1.5708e6/1e-10])

electrons_curr_ = np.concatenate([init_electrons[:7000, 1]*e*1.5708e6/1e-9, electrons_[:, 1]*e*1.5708e6/(2*1e-10)])
ions_curr_ = np.concatenate([init_ions[:7000, 1]*e*1.5708e6/1e-9, ions_[:, 1]*e*1.5708e6/(2*1e-10)])

electrons_curr__ = np.concatenate([init_electrons[:7000, 1]*e*1.5708e6/1e-9, electrons__[:, 1]*e*1.5708e6/(5*1e-10)])
ions_curr__ = np.concatenate([init_ions[:7000, 1]*e*1.5708e6/1e-9, ions__[:, 1]*e*1.5708e6/(5*1e-10)])

electrons_curr___ = np.concatenate([init_electrons[:7000, 1]*e*1.5708e6/1e-9, electrons___[:, 1]*e*1.5708e6/(5*1e-11)])
ions_curr___ = np.concatenate([init_ions[:7000, 1]*e*1.5708e6/1e-9, ions___[:, 1]*e*1.5708e6/(5*1e-11)])

fig, ax = plt.subplots()
fig1, ax1 = plt.subplots()

ax.plot(time, electrons_curr, label='1e-13')
ax.plot(time_, electrons_curr_, label='2e-13')
ax.plot(time__, electrons_curr__, label='5e-13')
ax.plot(time___, electrons_curr___, label='5e-14')

ax1.plot(timei, ions_curr, label='1e-13')
ax1.plot(time_i, ions_curr_, label='2e-13')
ax1.plot(time__i, ions_curr__, label='5e-13')
ax1.plot(time___i, ions_curr___, label='5e-14')

ax.set_title('Cathode current')
ax.set_xlabel('t, ns')
ax.set_ylabel('I, A')
ax.legend()
ax1.set_title('Anode current')
ax1.set_xlabel('t, ns')
ax1.set_ylabel('I, A')
ax1.legend()
```
