# Density

```python
Nz = 100
Nr = 50
dr = 5e-4
dz = 5e-4
coords_e = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_coords.txt")
coords_i = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/ions_final_coords.txt")
coords_e_del = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_coords_del.txt")
coords_i_del = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/ions_final_coords_del.txt")
dens_e = np.zeros((Nr, Nz))
dens_i = np.zeros((Nr, Nz))
dens_e_del = np.zeros((Nr, Nz))
dens_i_del = np.zeros((Nr, Nz))
nr = 0
nz = 0
for coords in coords_e:
    nz = int(coords[0] / dz)
    nr = int(coords[1] / dr)
    dens_e[nr][nz] += 1 #(np.pi * dz * dr * dr * (nr * nr - (nr - 1) * (nr - 1)))**(-1) * 157080
    dens_e_del[nr][nz] += 1 #(np.pi * dz * dr * dr * (nr * nr - (nr - 1) * (nr - 1)))**(-1) * 157080
for coords in coords_i:
    nz = int(coords[0] / dz)
    nr = int(coords[1] / dr)
    dens_i[nr][nz] += 1 #(np.pi * dz * dr * dr * (nr * nr - (nr - 1) * (nr - 1)))**(-1) * 157080
    dens_i_del[nr][nz] += 1 #(np.pi * dz * dr * dr * (nr * nr - (nr - 1) * (nr - 1)))**(-1) * 157080
```

```python
pth_e = ["long_calc_rho/B_0.1_sgm/1/electrons_Ntot.txt", "long_calc_rho/B_0.1_sgm/2/electrons_Ntot.txt",
         "long_calc_rho/B_0.1_sgm/3/electrons_Ntot.txt", "long_calc_rho/B_0.1_sgm/4/electrons_Ntot.txt",
         "long_calc_rho/B_0.1_sgm/5/electrons_Ntot.txt", "long_calc_rho/B_0.1_sgm/6_d_standart_dt/electrons_Ntot.txt"]
pth_i = ["long_calc_rho/B_0.1_sgm/1/ions_Ntot.txt", "long_calc_rho/B_0.1_sgm/2/ions_Ntot.txt", 
         "long_calc_rho/B_0.1_sgm/3/ions_Ntot.txt", "long_calc_rho/B_0.1_sgm/4/ions_Ntot.txt",
         "long_calc_rho/B_0.1_sgm/5/ions_Ntot.txt", "long_calc_rho/B_0.1_sgm/6_d_standart_dt/ions_Ntot.txt"]

electrons = []
ions = []
lengths = []
for j in range(len(pth_e)):
        electrons.append(np.genfromtxt(pth_e[j]))
        ions.append(np.genfromtxt(pth_i[j]))
        lengths.append(len(electrons[j]))
print(len(electrons))
time = []
k = 0
L = 0
for i in range(len(pth_e)):
        for j in range(lengths[i]):
                if i != 5:
                        time.append((electrons[i][j, 0] + 1000*L)*1e-3)
                elif i == 5:
                        time.append((electrons[i][j, 0]*1e-3 + 1000*L*1e-3))
                elif i == 5 and k >= 1e5:
                        time.append((electrons[i][j, 0]*1e-3 + 1000*(L + 1e5)*1e-3))
                k += 1
        k == 0
        L += lengths[i]

value_electrons = []
value_ions = []
for i in range(len(pth_e)):
        for j in range(lengths[i]):
                if i != 5:
                        value_electrons.append(electrons[i][j, 1])
                        value_ions.append(ions[i][j, 1])
                else:
                        value_electrons.append(electrons[i][j, 1]+ 2772236 - 277919)
                        value_ions.append(ions[i][j, 1] + 2325151 - 234018)


fig, ax = plt.subplots()

ax.plot(time,
        value_electrons,
        label = "electrons") #time
ax.plot(time,
        value_ions,
        label = "ions") #N

ax.set_ylabel('N')
ax.set_xlabel('t, ns')
ax.grid()
ax.set_title("Ntot")
ax.legend()
```

```python
ve_before = np.genfromtxt('long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_vel.txt')
vi_before = np.genfromtxt('long_calc_rho/B_0.1_sgm/final_after_1/ions_final_vel.txt')
coords_e_before = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_coords.txt")
coords_i_before = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/ions_final_coords.txt")

Nr = 50
Nz = 100

#Nr = int(Nr/5)
Nz = int(Nz/2)
    
Ee_before = np.zeros((Nr, Nz))
Ei_before = np.zeros((Nr, Nz))

nr = 0
nz = 0
dr, dz = 5e-4, 5e-4
e = 1.6e-19

me = 9.1e-31
mi = 100*me
matr_e_before = np.zeros((Nr, Nz))
matr_i_before = np.zeros((Nr, Nz))
i, j = 0, 0

for v in ve_before:
    nr = int(coords_e_before[i][1]/dr)
    nz = int(coords_e_before[i][0]/dz/2)
    Ee_before[nr][nz] += me * np.dot(v, v)/(2*e)
    matr_e_before[nr][nz] += 1
    i += 1
for nr_ in range(Nr):
    for nz_ in range(Nz):
        Ee_before[nr_][nz_] /= matr_e_before[nr_][nz_]
for v in vi_before:
    nr = int(coords_i_before[j][1]/dr)
    nz = int(coords_i_before[j][0]/dz/2)
    Ei_before[nr][nz] += mi * np.dot(v, v)/(2*e)
    matr_i_before[nr][nz] += 1
    j += 1
for nr_ in range(Nr):
    for nz_ in range(Nz):
        Ei_before[nr_][nz_] /= matr_i_before[nr_][nz_]
```

```python
ve = np.genfromtxt('long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_vels_del.txt')
vi = np.genfromtxt('long_calc_rho/B_0.1_sgm/final_after_1/ions_final_vels_del.txt')
coords_e = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/electrons_final_coords_del.txt")
coords_i = np.genfromtxt("long_calc_rho/B_0.1_sgm/final_after_1/ions_final_coords_del.txt")

Nr = 50
Nz = 100

#Nr = int(Nr/5)
Nz = int(Nz/2)
    
Ee = np.zeros((Nr, Nz))
Ei = np.zeros((Nr, Nz))


nr = 0
nz = 0
dr, dz = 5e-4, 5e-4
e = 1.6e-19

me = 9.1e-31
mi = 100*me
matr_e = np.zeros((Nr, Nz))
matr_i = np.zeros((Nr, Nz))
i, j = 0, 0

for v in ve:
    nr = int(coords_e[i][1]/dr)
    nz = int(coords_e[i][0]/dr/2)
    Ee[nr][nz] += me * np.dot(v, v)/(2*e)
    matr_e[nr][nz] += 1
    i += 1
for nr_ in range(Nr):
    for nz_ in range(Nz):
        Ee[nr_][nz_] /= matr_e[nr_][nz_]
for v in vi:
    nr = int(coords_i[j][1]/dr)
    nz = int(coords_i[j][0]/dz/2)
    Ei[nr][nz] += mi * np.dot(v, v)/(2*e)
    matr_i[nr][nz] += 1
    j += 1
for nr_ in range(Nr):
    for nz_ in range(Nz):
        Ei[nr_][nz_] /= matr_i[nr_][nz_]
        
        
fig, ax = plt.subplots()
ax.plot([i * 5e-2 for i in range(Nr)], Ee[:, 10], label = "electrons after delete")
ax.plot([i * 5e-2 for i in range(Nr)], Ee_before[:, 10], label = "electrons before delete")
ax.set_xlabel('$r, cm$')
ax.set_ylabel('$E, eV$')
ax.legend()

fig1, ax1 = plt.subplots()
ax1.plot([i * 5e-2 for i in range(Nr)], Ei[:, 10], label = "ions after delete")
ax1.plot([i * 5e-2 for i in range(Nr)], Ei_before[:, 10], label = "ions before delete")
ax1.set_xlabel('$r, cm$')
ax1.set_ylabel('$E, eV$')
ax1.legend()

fig2, ax2 = plt.subplots()
ax2.plot([i * 5e-2 for i in range(Nr)], matr_e_before[:, 10], label = "electrons before delete")
ax2.set_xlabel('$r, cm$')
ax2.set_ylabel('$Ntot$')
ax2.plot([i * 5e-2 for i in range(Nr)], matr_e[:, 10] /0.6, label = 'electrons after delete')
ax2.legend()

fig3, ax3 = plt.subplots()
ax3.plot([i * 5e-2 for i in range(Nr)], matr_i_before[:, 10], label = "ions before delete")
ax3.set_xlabel('$r, cm$')
ax3.set_ylabel('$Ntot$')
ax3.plot([i * 5e-2 for i in range(Nr)], matr_i[:, 10] /0.6 , label = 'ions after delete')
ax3.legend()
```
