```python
import numpy as np
from matplotlib import pyplot as plt
```

# Сделал в 10 раз болбше начальное число частиц с неизменным ptcls_per_macro = n_init * volume / Ntot

```python
ions = np.genfromtxt('ions_Ntot.txt')
electrons = np.genfromtxt('electrons_Ntot.txt')

time_e = electrons[:, 0] * 1e-3
time_i = ions[:, 0] * 1e-3

fig, ax = plt.subplots()

ax.plot(time_e, electrons[:, 1], label='electrons')
ax.plot(time_i, ions[:, 1], label='ions')
ax.set_xlabel('t, ns')
ax.set_ylabel('Ntot')
ax.legend()
```
